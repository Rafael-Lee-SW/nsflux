# Embedding Model & Configuration

```python
import torch
from transformers import AutoModel, AutoTokenizer

# Embedding Model
model = AutoModel.from_pretrained('BM-K/KoSimCSE-roberta-multitask',device_map="auto")
tokenizer = AutoTokenizer.from_pretrained('BM-K/KoSimCSE-roberta-multitask')
```

```python
# Json 파일인 경우
import json
# file_dir = "/home/user/바탕화면/RAG_AI/yeojun/1018/1018tablesummary_merged.json"
file_dir = "/home/user/바탕화면/RAG_AI/yeojun/1104/1104_NS_DB_old.json"


with open (file_dir,'r') as f:
    data=json.load(f)
```

```python
new_data = []
for i,d in enumerate(data):
    print(f'{i}번째 데이터')
    tmp = {
        "원본데이터":d['원본데이터'],
        "제목텍스트":d['제목텍스트'].replace("\n", " "),
        "수정일자":d['수정일자'],
        "요약데이터":d['요약데이터'].replace("\xa0"," ").replace("\x0b"," "),
        "청킹데이터":d['청킹데이터'].replace("\xa0"," ").replace("\x0b"," "),
        "출력데이터":d['출력데이터'],
    }
    new_data.append(tmp)
    
```

# Embedding & Savings

```python
# Json 파일인 경우
import re
import json
import numpy as np
from tqdm import tqdm

all_str_data = []
for i,d in tqdm(enumerate(new_data)):
    # 초기작업
    file_name = d['원본데이터']
    title = d['제목텍스트']
    date = d['수정일자']
    content = d['요약데이터']
    content_og = d['청킹데이터']
    content_vis = d['출력데이터']

    # 특수문자 전처리
    # content = re.sub(r'[\-\*\n]','',content)

    # 벡터화 진행
    inputs = tokenizer(content, padding=True, truncation=True, return_tensors="pt")
    embeddings, _ = model(**inputs, return_dict=False)
    vector = embeddings[:,0,:].detach().numpy().tolist() # (1, ?, embedding_dims)

    # 데이터 형태 저장
    structured_data = \
    {
        "file_name": f"{file_name}",
        "chunks": [
            {"chunk_id": i + 1, "title":title, "date":date, "type":"text", "text": content_og, "text_short": content, "vector": vector, "text_vis":content_vis}
            ]
    }
    all_str_data.append(structured_data)

# JSON 파일로 저장
with open('/home/user/바탕화면/RAG_AI/data/VectorDB/1104_NS_DB_old.json', 'w', encoding='utf-8') as json_file:
    data=json.dumps(all_str_data, ensure_ascii=False)
    json_file.write(data)
```
